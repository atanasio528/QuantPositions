```python
import pandas as pd
from sqlalchemy import create_engine

DB_USER = 'postgres'        
DB_PASSWORD = '0000' 
DB_HOST = 'localhost'        
DB_PORT = '5432'             
DB_NAME = 'quantpositions'  

engine = create_engine(f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}')

users_df = pd.read_sql('SELECT * FROM users', engine)
companies_df = pd.read_sql('SELECT * FROM companies', engine)
positions_df = pd.read_sql('SELECT * FROM positions', engine)
applied_df = pd.read_sql('SELECT * FROM applied', engine)

display(users_df)
display(companies_df)
display(positions_df)
display(applied_df)
```
